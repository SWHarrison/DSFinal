# How crappy is San Francisco?

Before coming here, many people question just how bad it could be compared to news stories.

### Let's find out!

```python
# Data taken on March 4th, 2019
# https://data.sfgov.org/City-Management-and-Ethics/Street-and-Sidewalk-Cleaning/h3eg-w3pj
# Importing data and sorting. Takes ~5 minutes with sort, 1 without.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("it's trying it's best")
data = pd.read_csv('Street_and_Sidewalk_Cleaning.csv')

#data.Opened = pd.to_datetime(data.Opened)

#ata.sort_values(by = 'Opened')

data
```

```python
#Counts of all request types
data['Request Type'].value_counts().plot(kind = 'bar')
```

```python
#Pie chart of counts
data['Neighborhood'].value_counts().plot(kind='pie', title = ' ')
```

```python
#Raw data of above 2 charts

data['Neighborhood'].value_counts()
```

```python
# Data for areas around MS

Nob = data[(data.Neighborhood == 'Lower Nob Hill')]
Tenderloin = data[(data.Neighborhood == 'Tenderloin')]
#Civic = data[(data.Neighborhood == 'Civic Center')]
MS_data = pd.concat([Nob,Tenderloin], ignore_index = True)
MS_data.Neighborhood.value_counts()
```

```python
# Counts of requests in Tenderloin

Tenderloin["Request Type"].value_counts().plot(kind = 'bar')
```

```python
# Above chart in pie form

Tenderloin["Request Type"].value_counts().plot(kind = 'pie')
```

```python
# Raw data for above 2 charts

Tenderloin["Request Type"].value_counts()
```

```python
# Counts of requests in Lower Nob Hill

Nob["Request Type"].value_counts().plot(kind = 'bar')
```

```python
# Aggregate of above 2 sections

MS_data["Request Type"].value_counts().plot(kind = 'bar')
```

```python
# In pie form

MS_data["Request Type"].value_counts().plot(kind = 'pie')
```

```python
# Raw data

MS_data['Request Type'].value_counts()
```
